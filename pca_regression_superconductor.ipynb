```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
import os
os.getcwd()
```

```python
#path0 = '/Users/Apple/Desktop/machine_learning/PCA/superconduct/'
supercond_tr = pd.read_csv('superconductor_train.csv')
#supercond_un = pd.read_csv(path0+'unique_m.csv')
```

```python
supercond_tr_x = supercond_tr.iloc[:,:81]
supercond_tr_y = supercond_tr.critical_temp
```

```python
supercond_tr_x.head()
```

```python
supercond_tr_y.head()
```

```python
supercond_tr_scaled = supercond_tr_x.apply(zscore)
```

```python
supercond_tr_scaled.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
pca.fit_transform(supercond_tr_scaled)
supercond_tr_pca = pca.fit_transform(supercond_tr_scaled)
```

```python
# Let us investigate the proportion 
# of variance explained and the elbow method
# pca.get_covariance()
pca.explained_variance_
# len(pca.explained_variance_)
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
# Let us investigate the performance of the models

from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor(max_depth=6)
DT.fit(supercond_tr_scaled,supercond_tr_y)
```

```python
rmse_without_pca = mean_squared_error(supercond_tr_y, DT.predict(supercond_tr_scaled))
rmse_without_pca
```

```python
from sklearn.tree import DecisionTreeRegressor
DT_pca = DecisionTreeRegressor(max_depth=6)
DT_pca.fit(supercond_tr_pca,supercond_tr_y)
```

```python
rmse_with_pca = mean_squared_error(supercond_tr_y, DT_pca.predict(supercond_tr_pca))
rmse_with_pca
```
